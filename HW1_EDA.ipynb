```python
df=df.drop('Unnamed: 0',1)
```

```python
df.describe()
```

```python
sns.pairplot(df)
```

As you can see there are positive trend between customer liftime value, Monthly Premium auto and Total claim ammount

```python
sns.catplot(x="Education", y="Customer_Lifetime_Value", data=df, kind="box", aspect=1.5)
plt.yscale('log')
plt.title("Boxplot for Customer_Lifetime_Value vs Education")
plt.show()
```

```python
sns.barplot(x="EmploymentStatus", y="Customer_Lifetime_Value", data=df)
plt.title("Boxplot for Customer_Lifetime_Value vs Education")
plt.show()
```

```python
data_num=df.select_dtypes(exclude="object")
```

```python
corrmat = data_num.corr()
sns.heatmap(corrmat, annot=True)
plt.show()
```
